```python
data['Gender'].fillna(0, inplace = True)

data['1st'].fillna(data['1st'].mean(),inplace = True) 
data['2nd'].fillna(data['2nd'].mean(),inplace = True) 
data['3rd'].fillna(data['3rd'].mean(),inplace = True) 
data['4th'].fillna(data['4th'].mean(),inplace = True) 
data['5th'].fillna(data['5th'].mean(),inplace = True) 


```

```python
data['Roll'] = data['Roll'].fillna(0).astype(int)
```

```python
data['Roll no.'].fillna(data['Roll no.'].mean(),inplace = True) 
```

```python
data.isnull().sum()
```

```python
data['College Code'] = data['College Code'].astype('str')
data['Roll no.'] = data['Roll no.'].astype('str')
data['Subject Code'] = data['Subject Code'].astype('str')
data['Student_id'] = data['Subject Code'] + data['Roll no.'] + data['College Code']
data['Student_id']
```

```python
data.drop(columns = ['Subject Code', 'Roll no.'], inplace = True)
```

```python
data['College Code'] = data['College Code'].astype(int)
```

```python
data.info()
```

```python
# I want to find a variable that have relationship with the 1st, 2nd, 3rd, 4th and 5th Column.

data['Avg']=(data['1st']+data['2nd']+data['3rd']+data['4th']+data['5th'])/5
```

```python
data
```

```python
#Select only revelent column  
data1 = data[['1st','2nd','3rd','4th','5th']]
data1
```

```python
x = data1
y = data.Avg
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.5,random_state = 23)
```

```python
x_train
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression 
LR=LinearRegression() #object Creation 
LR.fit(x_train,y_train)  #training of Linear Regression
y_predict=LR.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import r2_score , mean_squared_error, mean_absolute_error
r2_score = r2_score(y_test,y_predict)
r2_score
```

```python
import math 
print(mean_squared_error(y_test,y_predict))
print(math.sqrt(mean_squared_error(y_test,y_predict)))
```

```python
print(mean_absolute_error(y_test,y_predict))
```
